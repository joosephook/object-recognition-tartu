```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
submissions = [
    'joosep_submissions/1668260505/1668260505_test.csv',
    'joosep_submissions/1668261165/1668261165_test.csv',
     ]

scores = [ 0 for _ in submissions ]

predictions = [pd.read_csv(s) for s in submissions]
labels = pd.read_csv('labels.csv')


image_ids = pd.read_csv('test.csv')['image_id']
for image_id in image_ids:
    fig, ax = plt.subplots(1, 1)
    img = plt.imread('images/'+image_id)
    ax.imshow(img)
    ax.axis('off')
    plt.show()
    options = []
    for i, p in enumerate(predictions):
        l = p.loc[p.image_id == image_id, 'labels'].iloc[0]
        l = l.split(' ')
        ls = labels.loc[labels.label_id.isin(l), 'object'].tolist()
        ls = ' '.join(ls)
        print(i, ls)

    ans = input('>')
    bests = list(map(int, ans.strip().split(' ')))
    for i in bests:
        scores[i] += 1
    print(scores)

for s, p in zip(scores, predictions):
    print(s, p)
```
